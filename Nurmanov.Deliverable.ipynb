### Barplot

I created a barplot of room types. In order to see what type of room is most and leat popular in New York

```python
# Obtain unique values for room_type, 
# so it'd be easier to subset the dataset
df.room_type.unique()
```

```python
# defining figuresize and theme for the plot
plt.rcParams['figure.figsize'] = (20, 10)
plt.style.use('tableau-colorblind10')

# Subset by two dataframes
# Each subseted dataframe represents a type of room rented
# Private room
private = df[df['room_type'] == 'Private room']
private = private.replace('Private room', 1)
# Entire home or apartment
entire = df[df['room_type'] == 'Entire home/apt']
entire = entire.replace('Entire home/apt', 1)
# Shared room
shared = df[df['room_type'] == 'Shared room']
shared = shared.replace('Shared room', 1)

# Create data for the plot
# as dictionary
data = {"Private room": private['room_type'].sum(),
       "Entire home/apartment": entire['room_type'].sum(),
       "Shared room": shared['room_type'].sum()}

courses = list(data.keys())
values = list(data.values())


# creating the bar plot 
ax = plt.bar(courses, values, color =['blue','green', 'red'],  
        width = 0.5) 

# customizing the bar plot
plt.tick_params(labelsize=22)  
plt.title("Distribution of room types in New York", size=30, color='Black') 

plt.xticks(color = 'black', fontsize = 25)
plt.yticks(color = 'black', fontsize = 25)


# print the plot
plt.show()





```

## Geographical plot

On geographical plot of NYC we are able to the exact location of each entry and types that divided by color. 

```python
# Import data again

# This time I only choose choose only id, neighbourhood_type, room_type, price columns. 
# As well as adding latitude and longitude
geog_df = pd.read_csv("/Users/tnurmanov/Desktop/Semester 2/Data 506/archive/AB_NYC_2019.csv", usecols=(0,4,6,7,8,9))
# Rename neighbourhood_group to BoroName
geog_df.rename(columns = {'neighbourhood_group':'BoroName'}, inplace = True)
geog_df.head()
```

```python
# Import neccessary packages in order to create geographical plot
import geopandas as gpd
import geoplot as gplt
```

```python
# Remove the most extreme 1% prices,
# the most extreme .1% latitudes, &
# the most extreme .1% longitudes
geog_df = geog_df[(geog_df['price'] >= np.percentile(geog_df['price'], 0.5)) & 
 (geog_df['price'] <= np.percentile(geog_df['price'], 99.5)) & 
 (geog_df['latitude'] >= np.percentile(geog_df['latitude'], 0.05)) & 
 (geog_df['latitude'] < np.percentile(geog_df['latitude'], 99.95)) &
 (geog_df['longitude'] >= np.percentile(geog_df['longitude'], 0.05)) & 
 (geog_df['longitude'] <= np.percentile(geog_df['longitude'], 99.95))]
```

```python
from shapely.geometry import Point, Polygon

# import street map
street_map = gpd.read_file('/Users/tnurmanov/Desktop/Semester 2/Data 506/archive/Borough Boundaries/geo_export_d0ec678e-7d7a-40e0-8a35-dce39fafdc6a.shp')
```

```python
# designate coordinate system
crs = {'init':'epsg:4326'}
# zip x and y coordinates into single feature
geometry = [Point(xy) for xy in zip(geog_df['longitude'], geog_df['latitude'])]
# create GeoPandas dataframe
geo_frame = gpd.GeoDataFrame(geog_df,
 crs = crs,
 geometry = geometry)

geo_frame.head()
```

```python
# create figure and axes, assign to subplot
fig, ax = plt.subplots(figsize=(15,15))
# add .shp mapfile to axes
street_map.plot(ax=ax, alpha=0.4,color='gray')

palette = {'Entire home/apt':'blue', 'Private room':'green', 'Shared room':'red'}
cmap = matplotlib.colors.ListedColormap([palette[b] for b in geo_frame.room_type.unique()])

# add geodataframe to axes
# assign ‘price’ variable to represent coordinates on graph
# add legend
# make datapoints transparent using alpha
# assign size of points using markersize
geo_frame.plot(column='room_type',ax=ax,alpha=0.5, legend=True, markersize=8, cmap=cmap)

#,color='room_type', color_discrete_map={'Entire home/apt':'green', 'Private room':'blue', 'Shared room':'red'}

# add title to graph
plt.title('Types of apartments in NYC', fontsize=15,fontweight='bold')

# set latitiude and longitude boundaries for map display
plt.xlim(-74.30,-73.7)
plt.ylim( 40.5,40.95)
# show map
plt.show()
```

As we could see on barplot before, most popular type is entire house or apartment. And those kind of properties are mostly located on Manhattan. 

## Pie Chart

Let's see percentage proportions of all listings among each NYC borough

```python
# libraries required
import matplotlib.pyplot as plt

# defining figuresize for the plot
plt.rcParams['figure.figsize'] = (13, 8)

# Creating pie chart
# In order to get proportion of values and apply it on plot
# we used .value_counts() function, 
# that returns a total number of values under a certain condition
labels =  'Staten Island', 'Manhattan',  'Queens',  'Bronx',  'Brooklyn'

sizes = [df['neighbourhood_group'].value_counts()[4],
        df['neighbourhood_group'].value_counts()[0],
        df['neighbourhood_group'].value_counts()[2],
        df['neighbourhood_group'].value_counts()[3],
        df['neighbourhood_group'].value_counts()[1]]

explode = [0,0.1,0,0,0.1]


fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, explode = explode, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# This is just to get full percent proportions
df['neighbourhood_group'].value_counts(normalize=True)*100
```

Most properties of Airbnb are listed in Manhattan. Next is Brooklyn. And there're very few in Staten Island. Only 0.8% of all listings.

## Bar Chart

Comparison of boroughs' average prices

```python
# defining figuresize and theme for the plot
plt.rcParams['figure.figsize'] = (18, 8)
plt.style.use('ggplot')

# Subseting the data
manhattan = df[df['neighbourhood_group'] == 'Manhattan']
brooklyn = df[df['neighbourhood_group'] == 'Brooklyn']
queens = df[df['neighbourhood_group'] == 'Queens']
bronx = df[df['neighbourhood_group'] == 'Bronx']
si = df[df['neighbourhood_group'] == 'Staten Island']




data = {'Manhattan':manhattan['price'].mean(), 
        'Brooklyn':brooklyn['price'].mean(), 
        'Queens':queens['price'].mean(),  
        'Bronx':bronx['price'].mean(), 
        'Staten Island':si['price'].mean()} 

courses = list(data.keys()) 
values = list(data.values())  
  
# creating the bar plot 
plt.bar(courses, values, color ='green',  
        width = 0.4) 

# customizing the bar plot
plt.tick_params(labelsize=15)  
plt.ylabel("Average price", size=18) 
plt.title("Apartments' prices by average", size=20, color='k') 
plt.show()
```

As Manhattan is the most often area listed on Airbnb, it has the highest prices on average. Then are Brooklyn and Staten Island. Notably Staten Isalnd is the least often area listed on Airbnb, however it's price average is higher than in Queens and Bronx.

### Split-Apply-Combine

Here I create a function that returns five highest values of variable "price". From lower to higher/ Meaning the last value is the highest price of all New York City.

```python
def top(df, n=5, column='price'):
    return df.sort_values(by=column)[-n:]
top(df, n = 10)
```

Then I grouped by column 'neighbourhood_group'. Here are 5 the most expensive listings of each borough on Airbnb.

```python
df.groupby("neighbourhood_group").apply(top)
```
